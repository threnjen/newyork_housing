# Clean listing history

Function to find the second instance of Sold in a priceHistory event

```python
df_temp_history
```

```python
columns = df_temp_history.columns

#Find the second instance of Sold in a priceHistory/x/event
def find_last_history_start(x, columns, search_string):
    '''ARGUMENTS
    x = line
    columns = column headers'''
    for num in range (0,15):
        search = 'priceHistory/{}/event'.format(num)
        if x[search] == search_string:
            for num2 in range (num+1, 15):
                search2 = 'priceHistory/{}/event'.format(num2)
                if x[search2] == search_string:
                    return num2                  
            break

```

```python
#Fill all priceHistory events earlier and including 2nd Sold with NaN
for index in range(df_temp_history.shape[0]):
    x = df_temp_history.iloc[index]
    sold_num = find_last_history_start(x, columns, 'Sold')
    if sold_num == None:
        pass
    else:
        for num in range(sold_num,15):
            this_column = list(x.loc[columns.str.startswith('priceHistory/{}'.format(num))].index)
            for item in this_column:
                df_temp_history.iloc[index, df_temp_history.columns.get_loc(item)] = np.NaN
    list_num = find_last_history_start(x, columns, 'Listing removed')
    if list_num == None:
        pass
    else:
        for num in range(list_num,15):
            this_column = list(x.loc[columns.str.startswith('priceHistory/{}'.format(num))].index)
            for item in this_column:
                df_temp_history.iloc[index, df_temp_history.columns.get_loc(item)] = np.NaN
    print("Finished {}".format(index))
```

```python
# Count number of Price Change in a row and add to new column
def price_changes(line):
    changes = 0 # set local counter to 0
    
    for item in line: # look at each item in the line
        if item == 'Price change': # if the item is a Price change
            changes += 1 # increment counter by 1
        else: 
            continue

    return changes # return local counter
```

```python
# apply price change function to data frame
df_temp_history['price_changes'] = df_temp_history.apply(lambda x: price_changes(x), axis=1)
```

```python
#Find the earliest instance of Listed for Sale in a priceHistory/x/event
def find_earliest_price(x, columns):
    '''ARGUMENTS
    x = line
    columns = column headers'''
    for num in range (14, -1, -1):
        search = 'priceHistory/{}/event'.format(num)
        if x[search] == 'Listed for sale':
            number = x['priceHistory/{}/price'.format(num)]
            return number       

def find_earliest_date(x, columns):
    '''ARGUMENTS
    x = line
    columns = column headers'''
    for num in range (14, -1, -1):
        search = 'priceHistory/{}/event'.format(num)
        if x[search] == 'Listed for sale':
            number = x['priceHistory/{}/time'.format(num)]
            return number
```

```python


df_temp_history['initial_list_price'] = 0
df_temp_history['initial_time'] = 0

# apply price change function to data frame
df_temp_history['initial_list_price'] = df_temp_history.apply(lambda x: find_earliest_price(x, columns), axis=1)
df_temp_history['initial_time'] = df_temp_history.apply(lambda x: find_earliest_date(x, columns), axis=1)
```

```python
df_temp_history['sale_time'] = df_temp_history['priceHistory/0/time']
```

```python
df_temp_history.drop(df_temp_history.loc[:,df_temp_history.columns.str.startswith('priceHistory')],axis=1, inplace=True)
```

```python
df = pd.concat([df, df_temp_history], axis=1)
```
